```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(62)

# Parameters for the two modes
mode1_center = 40
mode1_std = 10
mode2_center = 75
mode2_std = 10

# Number of samples for each mode
n_samples_mode1 = 1000
n_samples_mode2 = 1000

# Generate the data for each mode
data_mode1 = np.random.normal(mode1_center, mode1_std, n_samples_mode1)
data_mode2 = np.random.normal(mode2_center, mode2_std, n_samples_mode2)

# Combine the data
data = np.concatenate([data_mode1, data_mode2])

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.hist(data, bins=100, edgecolor='k', alpha=0.7)
plt.title('Bimodal Distribution of Test Scores')
plt.xlabel('Test Score')
plt.ylabel('Frequency')
plt.show()
```
