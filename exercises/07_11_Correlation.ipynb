```python
import pandas as pd
from scipy.stats import shapiro

# Define the data
data = {
    'disposable_income': [
        605, 595, 578, 586, 612, 595, 578, 586, 612, 605,
        595, 578, 586, 612, 605, 595, 578, 586, 612, 605,
        595, 578, 586, 612, 605, 595, 578, 586, 612, 605,
        595, 578, 586, 612, 605, 595, 578, 586, 612, 605,
        595, 578, 586, 612, 605, 595, 578, 586, 612, 605
    ],
    'num_restaurants_visits': [
        5, 5, 4, 5, 6, 5, 4, 5, 6, 5,
        5, 4, 5, 6, 5, 5, 4, 5, 6, 5,
        5, 4, 5, 6, 5, 5, 4, 5, 6, 5,
        5, 4, 5, 6, 5, 5, 4, 5, 6, 5,
        5, 4, 5, 6, 5, 5, 4, 5, 6, 5
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Test for normality using Shapiro-Wilk test
stat_income, p_income = shapiro(df['disposable_income'])
stat_visits, p_visits = shapiro(df['num_restaurants_visits'])

# Print the results
print('Shapiro-Wilk Test for disposable_income:')
print('Statistic:', stat_income)
print('p-value:', p_income)

print('\nShapiro-Wilk Test for num_restaurants_visits:')
print('Statistic:', stat_visits)
print('p-value:', p_visits)

# Interpretation
alpha = 0.05
if p_income > alpha:
    print("\ndisposable_income looks Gaussian (fail to reject H0)")
else:
    print("\ndisposable_income does not look Gaussian (reject H0)")

if p_visits > alpha:
    print("num_restaurants_visits looks Gaussian (fail to reject H0)")
else:
    print("num_restaurants_visits does not look Gaussian (reject H0)")
```
