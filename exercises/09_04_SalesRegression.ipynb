```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
discounts = np.array([0, 5, 10, 15, 20, 25, 30]).reshape(-1, 1)
units_sold = np.array([10000, 12000, 12900, 16600, 18350, 19500, 22100])

# Create and train the model
model = LinearRegression()
model.fit(discounts, units_sold)

# Coefficients
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

# Predicting values
predicted_units_sold = model.predict(discounts)

# Plotting the data and the regression line
plt.scatter(discounts, units_sold, color='blue', label='Actual data')
plt.plot(discounts, predicted_units_sold, color='red', label='Regression line')
plt.xlabel('Discount Offered')
plt.ylabel('Units Sold')
plt.title('Linear Regression: Units Sold vs. Discount Offered')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm

# Data
discounts = np.array([0, 5, 10, 15, 20, 25, 30])
units_sold = np.array([10000, 12000, 12900, 16600, 18350, 19500, 22100])

# Reshape data for sklearn
discounts_reshaped = discounts.reshape(-1, 1)

# Create and train the model
model = LinearRegression()
model.fit(discounts_reshaped, units_sold)

# Coefficients
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

# Predicting values
predicted_units_sold = model.predict(discounts_reshaped)

# R-squared
r_squared = r2_score(units_sold, predicted_units_sold)
print(f"R-squared: {r_squared}")

# Using statsmodels to calculate p-value
discounts_with_intercept = sm.add_constant(discounts)  # Adds a constant term to the predictor
model_sm = sm.OLS(units_sold, discounts_with_intercept).fit()
p_value = model_sm.pvalues[1]  # p-value for the slope coefficient

print(f"P-value: {p_value}")

# Plotting the data and the regression line
plt.scatter(discounts, units_sold, color='blue', label='Actual data')
plt.plot(discounts, predicted_units_sold, color='red', label='Regression line')
plt.xlabel('Discount Offered')
plt.ylabel('Units Sold')
plt.title('Linear Regression: Units Sold vs. Discount Offered')
plt.legend()
plt.show()
```
